```python
import pandas as pd
import seaborn as sns 
import numpy as np
```

```python
countries = pd.read_csv("Data/countries of the world.csv",decimal=',')
```

```python
import pickle
```

```python
convCountry2Code = pickle.load(open("Data/countryToCode.pkl","rb"))
```

```python
countries["Country"] =countries["Country"].str.strip()
```

```python
countries["Code"] = countries["Country"].apply(lambda x: convCountry2Code[x] if x in convCountry2Code else "")
```

```python
combined_enforce = pd.read_pickle("Data/combined_enforce.pkl")
```

```python
combined_enforce = combined_enforce.reset_index()
```

```python
countries = countries.drop(["Region","Country"],axis=1)
```

```python
countries.columns = ["NATL_"+x if x != "Code" else x for x in countries.columns]
```

```python
countries.columns
```

```python
cbd = pd.merge(combined_enforce,countries,how="inner")
```

```python
cbd.set_index("Code")
```

```python
cbd.to_pickle("Data/allCombined.pkl")
```
