```python
def divide(x):
    if len(x)==1:
        return np.nan
    a = x[0]
    b = x[1]
        
    if a=="" or b=="":
        return 0
    else:
        if float(b)==0:
            return 0
        else:
            return float(a)/float(b)
```

```python
countries['Labour force participation (female/male pop. %)'] = countries['Labour force participation (female/male pop. %)'].str.split("/").apply(lambda x: divide(x))
```

```python
countries["Life expectancy at birth (females/males, years)"] =  countries['Life expectancy at birth (females/males, years)'].str.split("/").apply(lambda x: divide(x))
```

```python
countries["Population age distribution (0-14 / 60+ years, %)"] =  countries['Population age distribution (0-14 / 60+ years, %)'].str.split("/").apply(lambda x: divide(x))
```

```python
countries["International migrant stock (000/% of total pop.)"] =  countries['International migrant stock (000/% of total pop.)'].str.split("/").apply(lambda x: divide(x))
```

```python
countries['Education: Primary gross enrol. ratio (f/m per 100 pop.)'] =  countries['Education: Primary gross enrol. ratio (f/m per 100 pop.)'].str.split("/").apply(lambda x: divide(x))
```

```python
countries["Education: Secondary gross enrol. ratio (f/m per 100 pop.)"] =  countries['Education: Secondary gross enrol. ratio (f/m per 100 pop.)'].str.split("/").apply(lambda x: divide(x))
```

```python
countries["Education: Tertiary gross enrol. ratio (f/m per 100 pop.)"] =  countries['Education: Tertiary gross enrol. ratio (f/m per 100 pop.)'].str.split("/").apply(lambda x: divide(x))
```

```python
countries["Forested area (% of land area)"] = countries["Forested area (% of land area)"].str.split("/").apply(lambda x: divide(x))
```

```python
countries[['Pop. using improved drinking water (urban %)',"Pop. using improved drinking water (rural %)"]]= countries['Pop. using improved drinking water (urban/rural, %)'].replace("-99","nan/nan").str.split("/",expand=True).replace("", np.nan).astype(float)
```

```python
countries.drop("Pop. using improved drinking water (urban/rural, %)",axis=1,inplace=True)
```

```python
countries[['Pop. using improved sanitation facilities (urban %)',"Pop. using improved sanitation facilities (rural %)"]]= countries['Pop. using improved sanitation facilities (urban/rural, %)'].replace("-99","nan/nan").str.split("/",expand=True).replace("", np.nan).astype(float)
countries.drop("Pop. using improved sanitation facilities (urban/rural, %)",axis=1,inplace=True)
```

```python
countries["Energy supply per capita (Gigajoules)"] =  countries['Energy supply per capita (Gigajoules)'].str.split("/").apply(lambda x: divide(x))
```
